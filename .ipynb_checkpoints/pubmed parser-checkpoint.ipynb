```python
#IDS
idList = [elem.attrib for elem in article.xpath('//articleidlist/articleid[@idtype]')]
idDictList = []
for i in idList:
    for k,v in i.items():
        newPath = '//articleidlist/articleid[@idtype = "' + v + '"]'
        valsList = article.xpath(newPath)
        for val in valsList:
            idDict = {}
            idDict[v] = val.text_content()
            idDictList.append(idDict)          
```
